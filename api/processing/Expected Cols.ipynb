````python
def custom_sort(state_list):
    # Keep only strings that are exactly two alphabetical characters (e.g., 'CA', 'NY')
    two_letter_states = sorted([s for s in state_list if isinstance(s, str) and len(s) == 2 and s.isalpha()])
    # The rest of the list, sorted alphabetically
    other_states = sorted([s for s in state_list if s not in two_letter_states])
    return two_letter_states + other_states
````

````python
import re

def clean_key(key):
    # Convert to lowercase
    key = key.lower()
    # Replace non-alphabetical characters with '-'
    key = re.sub(r'[^a-z]', '-', key)
    return key
````

````python
print("# All non-numaric options")
for key in category_dict:
    ref = clean_key(key)
    print(f"* [{key}](#{ref}-field)")

for key in category_dict:
    print(f"\n\n## {key} field \n")
    print('```')
    for value in category_dict[key]:
        print(f"{value}")
    print('```')
````

````python
import re

def to_camel_case(s):
    # Split the string by spaces or underscores
    components = re.split(r'[\s_]+', s)
    # Return the first word in lowercase, and the rest capitalized
    return components[0].lower() + ''.join(x.title() for x in components[1:])
````

````python
validation_dict = {}
for key in category_dict:
    new_key = to_camel_case(key)
    validation_dict[new_key] = category_dict[key]
validation_dict['counselor'] = ["C9" if item == "C9'" else item for item in validation_dict['counselor']]
````

# Get Random Batch Data

````python
import pandas as pd
import random

def generate_csv(given_dict, n, filename="../default_batch.csv"):
    """
    Generates a CSV file with 'n' rows by selecting random values from the provided dictionary.
    Ensures unique Student IDs for each row.
    
    :param given_dict: Dictionary where keys are column names and values are lists of possible values.
    :param n: Number of rows to generate.
    :param filename: Name of the output CSV file (default is "../default_batch.csv").
    """
    sort_order_string = """state,country,gender,ethnicity,studentIDs,originSource,studentType,major,financialAidOfferedAmount,athlete,sport,raleyCollegeTagExists,recruitingTerritory,counselor,counselorIncomingTextCount,counselorOutgoingTextCount,phoneSuccessfulCount,phoneUnsuccessfulCount,phoneVoicemailCount,admittedStudentsDay,bisonDay,bisonDayAtTheWeekend,campusVisit,dallasBisonExclusive,footballVisit,golfVisit,oklahomaCityBisonExclusive,scholarsBisonDay,scholarsMixerAndBanquet,scholarshipInterview,scholarshipInterviewRegistration,softballVisit,trackVisit,tulsaBisonExclusive,volleyballVisit,eventsAttendedCount"""
    sort_order = sort_order_string.split(',')
    
    # Generate unique Student IDs
    student_id_range = range(1300000, 1600000)  # Ensure upper bound is included
    unique_student_ids = random.sample(student_id_range, n)  # Get 'n' unique IDs

    # Generate data while ensuring unique Student IDs
    data = {key: [random.choice(values) for _ in range(n)] for key, values in given_dict.items()}
    data['studentIDs'] = [f"{id_num:08d}" for id_num in unique_student_ids]  # Format as 8-digit numbers

    # Create DataFrame and sort columns
    df = pd.DataFrame(data)
    df = df[sort_order]  # Ensure columns are in correct order
    df.to_csv(filename, index=False)
    
    print(f"CSV file '{filename}' generated with {n} rows.")

````
